# Preparing data for machine learning model

```python
y = allData['ratingList'].iloc[1]
print(y)
```

```python
x = allData.customer_reviews.apply(lambda x: pd.Series(str(x).split("//")))
x.replace(np.nan, " ")
x.head(5)
```

```python
#data = select.drop([0,2,3,4,6,7,8,10,11,12,14,15,16,18,19,20,22,23,24,26,27,28,30,31,32,33,34,35,36,37], axis=1)
data = x.drop([1,2,3,5,6,7,9,10,11,13,14,15,17,18,19,21,22,23,25,26,27,29,30,31,33,34,35,37], axis=1)
data.head(20)
```

```python
data["merge"] = data[0].astype(str) + data[4].astype(str) + data[8].astype(str)+ data[12].astype(str)+ data[16].astype(str)+ data[20].astype(str)+ data[24].astype(str)+ data[28].astype(str)+ data[32].astype(str)+ data[36].astype(str)
data.head()
```

```python
data = data.drop([0,4,8,12,16,20,24,28,32,36], axis=1)
data.head()
```

```python
result = allData.join(data)
result.head()
```

```python
result = result.drop(['uniq_id', 'product_name', 'manufacturer', 'price', 'number_available_in_stock', 'number_of_reviews', 'number_of_answered_questions', 'amazon_category_and_sub_category', 'customers_who_bought_this_item_also_bought', 'description', 'product_information', 'product_description', 'items_customers_buy_after_viewing_this_item', 'customer_questions_and_answers', 'customer_reviews', 'sellers', 'ratingList'], axis=1)
result.head()
```

```python
dataFrame = result.groupby('reaction').apply(lambda x: x.sample(n=9982, replace = True)).reset_index(drop = True)
dataFrame.head()
```

```python
positiveDf = dataFrame.loc[dataFrame['reaction'] == 'positive']
positiveList = positiveDf['merge'].tolist()

negativeDf = dataFrame.loc[dataFrame['reaction'] == 'negative']
negativeList = negativeDf['merge'].tolist()
```

```python
positiveList_lower = [word.lower() for word in positiveList] 
```

```python
negativeList_lower = [word.lower() for word in negativeList]
```

```python
positiveListString = ' '.join([str(element) for element in positiveList_lower])  
negativeListString = ' '.join([str(element) for element in negativeList_lower])
```

```python
words = set(stopwords.words('english') + list(string.punctuation))
```

```python
tokenizer = WhitespaceTokenizer()
```

```python
filteredPositiveList = [w for w in tokenizer.tokenize(positiveListString) if w not in words] 

filteredNegativeList = [w for w in tokenizer.tokenize(negativeListString) if w not in words]
```

```python
filteredPositiveList2 = [w.strip(string.punctuation) for w in filteredPositiveList]
filteredNegativeList2 = [w.strip(string.punctuation) for w in filteredNegativeList]
```

```python
posWordFrequency = nltk.FreqDist(filteredPositiveList2) 
posWordFrequency
```

```python
negWordFrequency = nltk.FreqDist(filteredNegativeList2)
negWordFrequency
```

```python
def word_features(words):
     return dict([(word, True) for word in words.split()])
```

```python
positive_features = [(word_features(f), 'positive') for f in filteredPositiveList2]
negative_features = [(word_features(f), 'negative') for f in filteredNegativeList2]

labeledwords = positive_features + negative_features
```

```python
random.shuffle(labeledwords)
```

### Splitting Data into train and test

```python
#train_set, test_set = labeledwords[2000:], labeledwords[:500]
train_set, test_set = train_test_split(labeledwords, test_size=0.20)
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

### Checking the accuracy

```python
print(nltk.classify.accuracy(classifier, test_set))
```
